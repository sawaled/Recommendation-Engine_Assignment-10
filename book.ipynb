## 4. Model building

```python
df
```

```python
df_pivot = pd.pivot_table(data=df, values='Book_Rating', index='User_ID', columns= 'Book_Title').fillna(0)
df_pivot.index = df['User_ID'].unique()
df_pivot
```

```python
cosine_output = 1 - pairwise_distances(X = df_pivot , metric = 'cosine')
```

```python
cosine_output
```

```python
df_cosine = pd.DataFrame(data= cosine_output)
df_cosine.index = df['User_ID'].unique()
df_cosine.columns = df['User_ID'].unique()
df_cosine.head()
```
